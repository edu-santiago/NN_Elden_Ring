```python
target = 'target'
target_series = df[target]
feature_columns = df.drop(columns=[target]).columns

# Initialize a dictionary to store correlation results
correlations = {}

# Iterate through each feature column
for col in feature_columns:
    if df[col].dtype in [np.float64, np.int64]:  # Check if the column is continuous
        # Calculate point biserial correlation for continuous features
        corr, _ = pointbiserialr(df[col], target_series)
    else:
        # For categorical features, calculate the correlation using a suitable method
        # Here we use Cramer's V as an example for categorical-categorical correlation
        contingency_table = pd.crosstab(df[col], target_series)
        chi2 = np.sum((contingency_table - contingency_table.mean(axis=0))**2 / contingency_table.mean(axis=0))
        n = contingency_table.sum().sum()
        r, k = contingency_table.shape
        cramer_v = np.sqrt(chi2 / (n * (min(r, k) - 1)))
        corr = cramer_v
    
    correlations[col] = np.abs(corr)

# Convert the correlations dictionary to a DataFrame for better readability
correlation_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Absolute_Correlation'])
```
